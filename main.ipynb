<h1>Importing The Data</h1>

```python
import pandas as pd

# Import dataset
df = pd.read_csv("dataset/dataset.csv")
df.describe()
```

```python
df.head()
```

<h1>Data Preprocessing</h1>

```python
print("Shape")
print("Before dropping:", df.shape)

# Remove duplicates
df.drop_duplicates(inplace=True)

# Remove any rows that contain missing or null values 
df.dropna(inplace=True)

print("After dropping:", df.shape)
```

```python
from sklearn.preprocessing import StandardScaler

# Scaling the data
scaler = StandardScaler()
ndarray_scaled = scaler.fit_transform(df)

# Convert the scaled array back to a dataframe
df = pd.DataFrame(ndarray_scaled)

df.describe()
```

```python
df.head()
```

<h1>Clustering</h1>

```python
from sklearn.cluster import KMeans

from yellowbrick.cluster import KElbowVisualizer

# Perform elbow method
kmeans = KMeans(2, init="random", n_init=10, random_state=0)
visualizer = KElbowVisualizer(kmeans, k=(2,15), timings=False)
visualizer.fit(df)
visualizer.show()
```

```python
from sklearn.manifold import TSNE

import matplotlib.pyplot as plt

# Apply dimensionality reduction
tsne = TSNE(2, random_state=0)
df = tsne.fit_transform(df)

# Plot the result
plt.scatter(df[:,0], df[:,1])
plt.show()
```

```python
import numpy as np

# Perform k-means clustering
k = 6
kmeans = KMeans(k, init="random", n_init=10, random_state=0)
labels = kmeans.fit_predict(df)

# Plot clusters
for i in range(k):
    plt.scatter(df[np.asarray(labels == i), 0] , df[np.asarray(labels == i), 1] , label = i)

# Plot cluster centers
plt.scatter(kmeans.cluster_centers_[:,0] , kmeans.cluster_centers_[:,1] , s = 80, color='black')

plt.legend()
plt.show()
```
